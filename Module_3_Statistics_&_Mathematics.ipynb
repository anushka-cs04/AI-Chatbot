#Skewness
Skewness measures the symmetry of data.
1. Positive ➡ right-skewed
2. Negative ➡left-skewed
3. Zero ➡ Symmetry

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import skew

# ----------------------------
# Create Sample Dataset
# ----------------------------

np.random.seed(42)

# Normally distributed data
data = np.random.normal(loc=50, scale=10, size=500)

df = pd.DataFrame({
    "Marks": data
})


# ----------------------------
# Calculate Skewness
# ----------------------------

skew_value = skew(df["Marks"])

print("\nSkewness Value =", round(skew_value, 4))

# ----------------------------
# Interpretation
# ----------------------------

if skew_value > 0:
    print("Interpretation : Positively (Right) Skewed Distribution")
elif skew_value < 0:
    print("Interpretation : Negatively (Left) Skewed Distribution")
else:
    print("Interpretation : Perfectly Symmetric Distribution")

# ----------------------------
# Histogram
# ----------------------------

plt.figure(figsize=(8,5))

plt.hist(df["Marks"],
         bins=20,
         edgecolor="black")

plt.title("Histogram")
plt.xlabel("Marks")
plt.ylabel("Frequency")

plt.show()

# ----------------------------
# Density Plot
# ----------------------------

plt.figure(figsize=(8,5))

df["Marks"].plot(kind="density")

plt.title("Density Plot")
plt.xlabel("Marks")

plt.show()

# ----------------------------
# Box Plot
# ----------------------------

plt.figure(figsize=(6,4))

plt.boxplot(df["Marks"])

plt.title("Box Plot")

plt.show()
```
